```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import plotly
import plotly.express as px
```

```python
df = pd.read_csv("world_population.csv")
df
```

# Top 10 countries with highest population in 2022?

```python
data = df.nlargest(10, "2022 Population")[["Country/Territory", "2022 Population"]]
print(data.head(10))
```

```python
# Plotting
data.plot(kind='bar', x='Country/Territory', y='2022 Population', figsize=(10, 6))
plt.title('Countries with the Highest Population in 2022')
plt.xlabel('Country/Territory')
plt.ylabel('Population')
plt.show()
```

# Top 10 countries with low population in year 1970

```python
data=df.nsmallest(10,'1970 Population').groupby("Country/Territory")["1970 Population"].min()
data
```

```python
#plotting
data.plot(kind='bar', x='Country/Territory', y='1970 Population', figsize=(10, 6))
plt.title('Countries with the low Population in 1970')
plt.xlabel('Country/Territory')
plt.ylabel('Population')
plt.show()
```
